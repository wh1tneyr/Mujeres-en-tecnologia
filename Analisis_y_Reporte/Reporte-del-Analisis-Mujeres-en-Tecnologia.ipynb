### Sobre el Análisis
##### Este estudio busca explorar la participación de las mujeres en el sector tecnológico desde distintas dimensiones clave. A través de la comparación con sus pares masculinos, se pretende visibilizar las brechas existentes, identificar patrones de representación y destacar tanto los avances como los desafíos en materia de equidad de género dentro de la industria. Para ello, se abordan aspectos como la distribución por género en roles laborales y niveles jerárquicos, el uso y preferencia de lenguajes de programación, las diferencias en los métodos de aprendizaje, así como la experiencia y edad de inicio en la programación.

### Ejes de Análisis

1. Distribución de género en roles tecnológicos, niveles laborales e industrias
   - Analizar cómo se distribuyen mujeres y hombres en distintos puestos del sector tecnológico.
   - Observar la participación de cada género en diferentes niveles de seniority.
   - Evaluar la representación por género en las principales industrias que emplean perfiles tecnológicos.

2. Lenguajes de programación utilizados por género
   - Identificar los lenguajes de programación más utilizados por cada género.
   - Explorar la disposición a aprender nuevos lenguajes de programación.
   - Detectar posibles tendencias o preferencias según el género en el uso de determinados lenguajes.

3. Formación en programación y tecnología
   - Investigar los métodos de aprendizaje más comunes entre mujeres y hombres para adquirir habilidades en programación.
   - Analizar las edades de inicio en la programación según género, identificando posibles brechas de acceso temprano.

### 1. Distribución de género en roles tecnológicos, niveles laborales e industrias:

#### Distribución por género en distintos niveles laborales

El análisis muestra una mayor concentración de mujeres en niveles laborales junior, con aproximadamente un 21% frente al 18% de los hombres. Sin embargo, la brecha se amplía al observar la participación femenina en niveles senior, donde solo el 13% de las mujeres alcanzan estas posiciones frente al 25% de los hombres. Esta diferencia puede estar vinculada a factores como las brechas salariales y las limitadas oportunidades de ascenso. A pesar de su buena representación en etapas iniciales, las mujeres enfrentan barreras para acceder a puestos de liderazgo. Este hallazgo refuerza la importancia de impulsar estrategias que promuevan la equidad en el crecimiento profesional.

#### Distribución por género en diversos roles tecnológicos

En cuanto a los roles dentro del sector, se observa una mayor representación masculina en los cargos más técnicos. Los desarrolladores backend, fullstack y de software presentan una diferencia de representación masculina de entre 2% y 6%. En contraste, roles como frontend developer y data analyst muestran una ligera preferencia femenina, aunque las diferencias son marginales (alrededor del 1%). Aunque las mujeres han logrado ciertos avances en áreas como frontend y análisis de datos, los roles más especializados o vinculados a la infraestructura tecnológica siguen siendo dominados por hombres.

#### Participación por Industria

Al analizar la presencia por industria, las mujeres tienen una participación del 41% en el sector TI frente al 39% de los hombres. En educación, las mujeres también superan a los hombres (16% frente a 13%), mientras que en servicios financieros la situación es inversa: 9% de participación masculina frente al 6% femenina. La mayor presencia femenina en sectores como educación podría estar relacionada con una afinidad por entornos colaborativos o de impacto social, mientras que la menor participación en industrias como finanzas sugiere barreras o menor atracción hacia áreas históricamente masculinizadas.

### 2. Lenguajes de programación utilizados por género

#### Lenguajes de programación: los más usados y las preferencias por género

En cuanto a los lenguajes de programación, Java es el más utilizado por las mujeres, superando levemente a los hombres. Sin embargo, lenguajes como JavaScript y Python tienen una mayor presencia masculina, lo que refleja su dominio en roles técnicos altamente especializados. A pesar de ello, un 40% de las mujeres expresó interés en aprender Python, lo que indica una fuerte intención de especializarse en lenguajes versátiles y en alta demanda, como los que se utilizan en análisis de datos, automatización y aprendizaje automático.

### 3. Formación en programación y tecnología

#### Formación en programación: estudio normal vs. autodidactismo

El 76% de las mujeres adquiere sus habilidades en programación a través de la educación universitaria, mientras que el 50% lo hace de manera autodidacta. En cambio, el 74% de los hombres afirma haber aprendido por cuenta propia. Esta diferencia podría reflejar un acceso desigual a recursos informales, o bien, una mayor necesidad por parte de las mujeres de validar sus conocimientos mediante credenciales formales en un entorno que aún presenta sesgos de género.

####  Edad de inicio en la programación

La mayoría de las mujeres comienza a programar entre los 16 y 20 años, un rango similar al de los hombres. No obstante, la representación femenina en edades más tempranas (11–15 años) es menor, lo que podría estar influenciado por estereotipos de género o barreras culturales que dificultan el acceso temprano a experiencias tecnológicas. Aunque este dato refleja una entrada más tardía al sector, el crecimiento progresivo en los rangos de edad posteriores resulta alentador.

### Conclusión y reflexión final

 En resumen, aunque se evidencian avances en la participación femenina en el ámbito tecnológico, aún persisten desafíos significativos. Las mujeres siguen estando subrepresentadas en niveles senior y en roles altamente técnicos, lo que destaca la necesidad de implementar políticas inclusivas, programas de mentoría y estructuras de apoyo que faciliten su desarrollo y liderazgo.

Además, las diferencias en los métodos de formación revelan posibles inequidades en el acceso a recursos y oportunidades educativas. Fomentar el acceso temprano a la tecnología, visibilizar referentes femeninos y promover iniciativas que valoren la diversidad son pasos fundamentales para construir una industria más equitativa y representativa.
